# Основні методи списків в Python :

append(x) — додає елемент в кінець списку

extend(iterable) — додає всі елементи іншого списку (або ітератора)

insert(i, x) — вставляє елемент на вказану позицію

remove(x) — видаляє перше входження елемента

pop([i]) — видаляє елемент за індексом і повертає його

clear() — видаляє всі елементи списку

index(x[, start[, end]]) — повертає індекс першого входження елемента

count(x) — рахує кількість входжень елемента

sort(key=None, reverse=False) — сортує список

reverse() — розвертає елементи списку

copy() — повертає копію списку

# Основні методи для роботи з множинами (set) в Python:

add() — додати елемент.

remove() — видалити елемент (викидає помилку, якщо елемента немає).

discard() — видалити елемент (без помилки, якщо елемента немає).

pop() — видалити і повернути довільний елемент.

clear() — очистити множину (залишити порожню).

update() — додати кілька елементів одразу (з іншого списку, множини тощо).

union() — об'єднання множин.

intersection() — перетин множин.

difference() — різниця множин.

symmetric_difference() — симетрична різниця множин.

issubset() — перевірити, чи є підмножиною.

issuperset() — перевірити, чи є надмножиною.

isdisjoint() — перевірити, чи множини не мають спільних елементів.

copy() — створити копію множини.

# Основні методи для роботи з кортежами (tuple) в Python:

count(x) — підрахувати кількість елементів x у кортежі.

index(x) — знайти індекс першого входження елемента x.



# Основні методи для роботи зі словниками в Python.

```
dict(): Створення нового словника.
dict.keys(): Повертає список ключів у словнику.
dict.values(): Повертає список значень у словнику.
dict.items(): Повертає список пар (ключ, значення) у словнику.
dict.get(key): Повертає значення для зазначеного ключа, якщо ключ існує; інакше повертає None або значення за замовчуванням.
dict.setdefault(key, default): Повертає значення для зазначеного ключа, якщо ключ існує; інакше добавляє ключ з заданим значенням за замовчуванням.
dict.update(): Оновлює словник, додаючи пари ключ-значення з іншого словника або з ітерабельного об'єкта.
dict.pop(key): Видаляє ключ і повертає його значення. Якщо ключ не існує, виникає помилка KeyError.
dict.popitem(): Видаляє і повертає останню добавлену пару (ключ, значення) зі словника.
dict.clear(): Очищає словник, усунувши всі пари ключ-значення.
dict.copy(): Повертає поверхневу копію словника.
```

# Завдання 4: Перетин множин

Зробіть  дві множини. Знайдіть спільні елементи між ними

````python
# Завдання 4: Перетин множин
# Зробіть дві множини. Знайдіть спільні елементи між ними
# Перетин множин — це математична операція, яка визначає спільні елементи двох або більше множин.
# В результаті операції перетину формується нова множина, що містить тільки ті елементи, які є присутніми в усіх множинах.
# Створення двох множин

set1 = set(range(1, 6))  # Множина {1, 2, 3, 4, 5}. З допомогою set(range() спробувати створити множину не вручну
set2 = set([4, 5, 6, 7])  # Множина {4, 5, 6, 7} (з списку)

# Знаходження спільних елементів
intersection = set1.intersection(set2)   # Перетин множин
print("Спільні елементи:", intersection)

# альтернативний метод
intersection = set1 & set2               # Перетин множин
print("Спільні елементи2:", intersection)
````

# Завдання 7: Комбінації елементів

Зробіть   два списки. Знайдіть всі можливі комбінації пар елементів між ними

````python
# Завдання 7: Комбінації елементів
# Зробіть   два списки. Знайдіть всі можливі комбінації пар елементів між ними
'''Бібліотека itertools
Основні функції з itertools:

product

Картезіанський добуток (перебір всіх комбінацій). '''

import itertools           # спроба використати бібліотеки itertools (функція product() з цієї бібліотеки дозволяє створити декартовий добуток, тобто всі можливі комбінації.)

list1 = ['A', 'B', 'C']    # Два довільні списки
list2 = [1, 2, 3]

combinations = list(itertools.product(list1, list2))  # Знаходження всіх можливих комбінацій

print("Всі можливі комбінації пар елементів:")        # Виведення результату
for pair in combinations:
    print(pair)
````

# Завдання 8: Видалення елемента з множини

# Завдання 9: Сортування словника за значеннями

Зробіть  словник товарів і їхніх цін. Виведіть словник, відсортований за ціною

````python
# Завдання 9: Сортування словника за значеннями
# Зробіть  словник товарів і їхніх цін. Виведіть словник, відсортований за ціною

# Словник товарів і їхніх цін
products = {
    "apple": 0.5,
    "banana": 0.3,
    "orange": 0.8,
    "mango": 1.2,
    "grapes": 2.0
}

def get_price(item):                                            # def — це скорочення від define (визначити). Воно використовується для створення (визначення) функції.
    return item[1]
sorted_products = dict(sorted(products.items(), key=get_price)) # Сортує ці пари за значенням (item[1] — це ціна).
print(sorted_products)
for key, value in products.items():                             # вивести красивий список
  print(key, ":", value)
````

# Завдання 10: Конкатенація списків

Зробіть  два списки. Об'єднайте їх в один список.

````python
# Завдання 10: Конкатенація списків
# Зробіть  два списки. Об'єднайте їх в один список.

# Використати оператор +
list1 = [1, 2, 3]
list2 = [4, 5, 6]

combined_list = list1 + list2                # Конкатенація списків

print("Об'єднаний список №1:", combined_list)   # Виведення результату
````

````python
# Використати метод .extend()
list1 = [1, 2, 3]
list2 = [4, 5, 6]

list1.extend(list2)    # Додаємо елементи другого списку

# Виведення результату
print("Об'єднаний список №2:", list1)
````

````python
import itertools

# Використати itertools.chain()
list1 = [1, 2, 3]
list2 = [4, 5, 6]

combined_list = list(itertools.chain(list1, list2))   # Конкатенація списків

print("Об'єднаний список №3:", combined_list)         # Виведення результату
````

# Додаткове завдання

Є список (треба його задати), що містить елементи різних типів даних (числа, строки, списки, тощо). Розділіть цей список на окремі списки за типами даних

````python
# Задання початкового списку з елементами різних типів даних


mixed_list = [1, "hello", [1, 2, 3], 4.5, "world", (5, 6), 3, ["another", "list"], None, {1: "one", 2: "two"}]

# Беремо всі типи даних які знаємо і створимо категорії(по типам даних):
numbers = []
strings = []
lists = []
tuples = []
dictionaries = []
others = []

# Розділимо mixed_list за категоріями вище і, згідно типів даних, заповнимо категорії, які створили вище
for item in mixed_list:
    if type(item) == int or type(item) == float:  # Перевірка на числа
        numbers.append(item)
    elif type(item) == str:  # Перевірка на рядки
        strings.append(item)
    elif type(item) == list:  # Перевірка на списки
        lists.append(item)
    elif type(item) == tuple:  # Перевірка на кортежі
        tuples.append(item)
    elif type(item) == dict:  # Перевірка на словники
        dictionaries.append(item)
    else:  # Інші типи даних
        others.append(item)

# Виведення результатів
print("Числа:", numbers)
print("Рядки:", strings)
print("Списки:", lists)
print("Кортежі:", tuples)
print("Словники:", dictionaries)
print("Інші:", others)
````
